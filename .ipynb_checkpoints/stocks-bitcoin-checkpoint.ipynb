```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
import pandas_datareader as pdr
import datetime as dt
import numpy as np
 
```

```python
btc = pd.read_csv('BTC-2.csv',thousands=',')
stocks = pd.read_csv('Nasdaq.csv',thousands=',')
```

```python
btc['Price'].astype(int).corr(stocks['Price'].astype(int))
```

```python
start = dt.datetime(2020, 1, 1)
def visualize_correlation(ticker1, ticker2):
    df = pdr.get_data_yahoo([ticker1, ticker2], start)
    df = df['Adj Close']
    df = df/df.iloc[0]
    fig, ax = plt.subplots()
    df.plot(ax=ax)
    
visualize_correlation("AAPL", "NFLX")
```
